```python
# 5. Set the correct paths (edit DATASET_NAME to match your kaggle dataset folder)
import os

DATASET_NAME = os.listdir('/kaggle/input')[0]  # auto-detect dataset folder
BASE = f'/kaggle/input/{DATASET_NAME}'

# Find HAI and HIEND dirs
subdirs = os.listdir(BASE)
print('Subdirs:', subdirs)

HAI_DIR   = next(os.path.join(BASE, d) for d in subdirs if 'haiend' not in d.lower() and 'hai' in d.lower())
HIEND_DIR = next(os.path.join(BASE, d) for d in subdirs if 'haiend' in d.lower())

print('HAI_DIR  :', HAI_DIR)
print('HIEND_DIR:', HIEND_DIR)

os.environ['HAI_DIR']   = HAI_DIR
os.environ['HIEND_DIR'] = HIEND_DIR
```
